```python
from litellm import completion
import os
from pprint import pprint
import json
from dotenv import load_dotenv
load_dotenv()  # Load environment variables from .env file

models = json.load(open("backend/llm_prices.json"))

for model in models["prices"]:
    print(f"Testing model: {model['id']}")

    try:
        response = completion(
            model="vertex_ai/" + model["id"],
            messages=[{ "content": "Hello, how are you?","role": "user"}],
            vertex_ai_location=os.environ.get("VERTEX_REGION"),
            vertex_ai_project=os.environ.get("VERTEX_PROJECT_ID"),
        )
        print("Response for " + model["id"] + ":\n" + response.choices[0].message.content)
    except Exception as e:
        print(f"Error testing model {model['id']}: {e}")

```
